```python
import pandas as pd
```

```python
#resource files 
school_data_to_load = "../Resources/schools_complete.csv"
student_data_to_load = "../Resources/students_complete.csv"
```

```python
#reading the files 
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
school_data.head
```

```python
student_data.head
```

```python
# creating a single set of data
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

```python
#Counts
school_count = len(school_data["School ID"])

students_count = len(student_data["Student ID"])
```

```python
#BUDGET
total_budget = school_data["budget"].sum()
```

```python
#Average scores
avg_math_score = student_data['math_score'].mean()
avg_reading_score = student_data['reading_score'].mean()
```

```python
#Percent passing 
perc_students_passing_math = len(student_data[(student_data["math_score"] >= 70)]) * 100 / students_count
perc_students_passing_reading = len(student_data[(student_data["reading_score"] >= 70)]) * 100 / students_count
perc_students_passing_math_and_reading = len(student_data.loc[(student_data["math_score"] >= 70) & \
                                                              (student_data["reading_score"] >= 70)]) * 100 / students_count
```

```python
#District Information 
district_summary_df = pd.DataFrame({"Total Schools": [school_count],
                              "Total Students": students_count,
                              "Total Budget": total_budget,
                              "Average Math Score": avg_math_score,
                              "Average Reading Score": avg_reading_score,
                              "% Passing Math": perc_students_passing_math,
                              "% Passing Reading": perc_students_passing_reading,    
                              "% Overall Passing": perc_students_passing_math_and_reading})
district_summary_df.head()
```

```python
#School Ifo
school_data_group = school_data.groupby(['school_name'])
school_data_group_df = school_data_group.head(len(school_data_group))
school_data_group_df = school_data_group_df[['school_name', 'type', 'size', 'budget']]
school_data_group_df = school_data_group_df.rename(columns={
    #'school_name': 'School Name',
    'type': 'School Type',
    'size': 'Total Students',
    'budget': 'Total School Budget'})
school_data_group_df['Per Student Budget'] = school_data_group_df['Total School Budget'] / \
                                             school_data_group_df['Total Students']
school_data_group_df
```

```python
#student info
student_data_group = student_data.groupby(['school_name'])
student_data_group = student_data_group[['school_name', 'reading_score', 'math_score']]
student_data_group_mean_df = student_data_group.mean()
student_data_group_mean_df = student_data_group_mean_df.rename(columns={
    'reading_score': 'Average Reading Score',
    'math_score': 'Average Math Score'})
student_data_group_mean_df
```

```python
#Students Pass info for Math 
student_data_math_passing =  student_data[(student_data['math_score'] >= 70)]
student_data_math_passing_df = student_data_math_passing.groupby(['school_name'])
student_data_math_passing_df = student_data_math_passing_df.count()
student_data_math_passing_df = student_data_math_passing_df[['math_score']]
student_data_math_passing_df = student_data_math_passing_df.rename(columns={
    'math_score': 'Total Students Passing Math'
})
student_data_math_passing_df.head()
```

```python
#Students Passing Info for Reading 
student_data_reading_passing =  student_data[(student_data["reading_score"] >= 70)]
student_data_reading_passing_df = student_data_reading_passing.groupby(['school_name'])
student_data_reading_passing_df = student_data_reading_passing_df.count()
student_data_reading_passing_df = student_data_reading_passing_df[['reading_score']]
student_data_reading_passing_df = student_data_reading_passing_df.rename(columns={
    'reading_score': 'Total Students Passing Reading'
})
student_data_reading_passing_df.head()
```

```python
#Combines info 
student_data_math_and_reading_passing = student_data.loc[(student_data['math_score'] >= 70) & \
                                                              (student_data['reading_score'] >= 70)]
student_data_math_and_reading_passing_df = student_data_math_and_reading_passing.groupby(['school_name'])
student_data_math_and_reading_passing_df = student_data_math_and_reading_passing_df.count()
student_data_math_and_reading_passing_df = student_data_math_and_reading_passing_df[['Student ID']]
student_data_math_and_reading_passing_df = student_data_math_and_reading_passing_df.rename(columns = {
    'Student ID': 'Total Students Passing Math and Reading'
})
student_data_math_and_reading_passing_df.head()
```

```python
#Schools Summary 
school_summary1 = pd.merge(school_data_group_df, student_data_group_mean_df, on=['school_name'])
school_summary2 = pd.merge(school_summary1, student_data_math_passing_df, on=['school_name'])
school_summary3 = pd.merge(school_summary2, student_data_reading_passing_df, on=['school_name'])
school_summary = pd.merge(school_summary3, student_data_math_and_reading_passing_df, on=['school_name'])
school_summary['% Passing Math'] = school_summary['Total Students Passing Math'] * 100/ school_summary['Total Students']
school_summary['% Passing Reading'] = school_summary['Total Students Passing Reading'] * 100/ school_summary['Total Students']
school_summary['% Overall Passing'] = school_summary['Total Students Passing Math and Reading'] * 100/ school_summary['Total Students']
school_summary = school_summary[['school_name', 'School Type', 'Total Students', 'Total School Budget',
       'Per Student Budget', 'Average Reading Score', 'Average Math Score', '% Passing Math',
       '% Passing Reading', '% Overall Passing']]
school_summary = school_summary.rename(columns={
    'school_name': 'School Name'})
# Formatting
school_summary["Total Students"] = school_summary["Total Students"].map(
    "{:,}".format)
school_summary["Total School Budget"] = school_summary["Total School Budget"].map(
    "${:,.2f}".format)
school_summary["Average Math Score"] = school_summary["Average Math Score"].map(
    "{:,.2f}".format)
school_summary["Average Reading Score"] = school_summary["Average Reading Score"].map(
    "{:,.2f}".format)
school_summary["% Passing Math"] = school_summary["% Passing Math"].map(
    "{:,.2f}".format)
school_summary["% Passing Reading"] = school_summary["% Passing Reading"].map(
    "{:,.2f}".format)
school_summary["% Overall Passing"] = school_summary["% Overall Passing"].map(
    "{:,.2f}".format)

school_summary
school_summary_pd = pd.DataFrame(school_summary)
school_summary_pd = school_summary_pd.set_index('School Name')
school_summary_pd
```

```python
#Lowest Schools
school_summary_bottom5_overall_passing = school_summary_pd.sort_values("% Overall Passing")
school_summary_bottom5_overall_passing.head(5)
```

```python
#Best Schools 
school_summary_top5_overall_passing = school_summary_pd.sort_values("% Overall Passing", ascending=False)
school_summary_top5_overall_passing.head(5)
```

```python
#Grade level Math 
student_data_df = student_data[['school_name', 'grade', 'math_score']]
school_9th_grade_math = student_data_df[(student_data_df['grade'] == '9th')]
school_9th_grade_math = school_9th_grade_math.groupby(['school_name']).mean()
school_9th_grade_math = school_9th_grade_math.rename(columns={
    'math_score': '9th'})

school_10th_grade_math = student_data_df[(student_data_df['grade'] == '10th')]
school_10th_grade_math = school_10th_grade_math.groupby(['school_name']).mean()
school_10th_grade_math = school_10th_grade_math.rename(columns={
    'math_score': '10th'})

school_11th_grade_math = student_data_df[(student_data_df['grade'] == '11th')]
school_11th_grade_math = school_11th_grade_math.groupby(['school_name']).mean()
school_11th_grade_math = school_11th_grade_math.rename(columns={
    'math_score': '11th'})

school_12th_grade_math = student_data_df[(student_data_df['grade'] == '12th')]
school_12th_grade_math = school_12th_grade_math.groupby(['school_name']).mean()
school_12th_grade_math = school_12th_grade_math.rename(columns={
    'math_score': '12th'})

math_score1 = pd.merge(school_9th_grade_math, school_10th_grade_math, on=['school_name'])
math_score2 = pd.merge(math_score1, school_11th_grade_math, on=['school_name'])
math_score = pd.merge(math_score2, school_12th_grade_math, on=['school_name'])
math_score["9th"] = math_score["9th"].map(
    "{:,.2f}".format)
math_score["10th"] = math_score["10th"].map(
    "{:,.2f}".format)
math_score["11th"] = math_score["11th"].map(
    "{:,.2f}".format)
math_score["12th"] = math_score["12th"].map(
    "{:,.2f}".format)
math_score
```

```python
#Grade level Reading 
student_data_df = student_data[['school_name', 'grade', 'reading_score']]
school_9th_grade_reading = student_data_df[(student_data_df['grade'] == '9th')]
school_9th_grade_reading = school_9th_grade_reading.groupby(['school_name']).mean()
school_9th_grade_reading = school_9th_grade_reading.rename(columns={
    'reading_score': '9th'})

school_10th_grade_reading = student_data_df[(student_data_df['grade'] == '10th')]
school_10th_grade_reading = school_10th_grade_reading.groupby(['school_name']).mean()
school_10th_grade_reading = school_10th_grade_reading.rename(columns={
    'reading_score': '10th'})

school_11th_grade_reading = student_data_df[(student_data_df['grade'] == '11th')]
school_11th_grade_reading = school_11th_grade_reading.groupby(['school_name']).mean()
school_11th_grade_reading = school_11th_grade_reading.rename(columns={
    'reading_score': '11th'})

school_12th_grade_reading = student_data_df[(student_data_df['grade'] == '12th')]
school_12th_grade_reading = school_12th_grade_reading.groupby(['school_name']).mean()
school_12th_grade_reading = school_12th_grade_reading.rename(columns={
    'reading_score': '12th'})

reading_score1 = pd.merge(school_9th_grade_reading, school_10th_grade_reading, on=['school_name'])
reading_score2 = pd.merge(reading_score1, school_11th_grade_reading, on=['school_name'])
reading_score = pd.merge(reading_score2, school_12th_grade_reading, on=['school_name'])
reading_score["9th"] = reading_score["9th"].map(
    "{:,.2f}".format)
reading_score["10th"] = reading_score["10th"].map(
    "{:,.2f}".format)
reading_score["11th"] = reading_score["11th"].map(
    "{:,.2f}".format)
reading_score["12th"] = reading_score["12th"].map(
    "{:,.2f}".format)
reading_score
```

```python
#School Spending 
scores_school_spending = school_summary
scores_school_spending['Average Reading Score'] = scores_school_spending['Average Reading Score'].astype("float64")
scores_school_spending['Average Math Score'] = scores_school_spending['Average Math Score'].astype("float64")
scores_school_spending['% Passing Math'] = scores_school_spending['% Passing Math'].astype("float64")
scores_school_spending['% Passing Reading'] = scores_school_spending['% Passing Reading'].astype("float64")
scores_school_spending['% Overall Passing'] = scores_school_spending['% Overall Passing'].astype("float64")
spending_bins = [0, 584, 629, 644, 680]
spending_labels = ['<$585', '$585-630', '$630-645', '$645-680']

scores_school_spending['Spending Ranges (Per Student)'] = pd.cut(scores_school_spending['Per Student Budget'], spending_bins, labels=spending_labels)

scores_school_spending_group = scores_school_spending.groupby('Spending Ranges (Per Student)')

scores_school_spending_group[['Average Math Score', 'Average Reading Score', '% Passing Math', '% Passing Reading', '% Overall Passing']].mean()
```

```python
#School size scores 
scores_school_size = school_summary.copy()
scores_school_size['Total Students'] = scores_school_size['Total Students'].str.replace(',', '')
scores_school_size['Total Students'] = scores_school_size['Total Students'].astype("int")
scores_school_size['Average Reading Score'] = scores_school_size['Average Reading Score'].astype("float64")
scores_school_size['Average Math Score'] = scores_school_size['Average Math Score'].astype("float64")
scores_school_size['% Passing Math'] = scores_school_size['% Passing Math'].astype("float64")
scores_school_size['% Passing Reading'] = scores_school_size['% Passing Reading'].astype("float64")
scores_school_size['% Overall Passing'] = scores_school_size['% Overall Passing'].astype("float64")

size_bins = [0, 999, 1999, 5000]
size_labels = ["Small (<1000)", 'Medium (1000-2000)', 'Large (2000-5000)']


scores_school_size['School Size'] = pd.cut(scores_school_size['Total Students'], size_bins, labels=size_labels)

scores_school_size_group = scores_school_size.groupby('School Size')

scores_school_size_group[['Average Math Score', 'Average Reading Score', '% Passing Math', '% Passing Reading', '% Overall Passing']].mean()
```

```python
#School Types 
scores_school_type = school_summary.copy()
scores_school_type['Average Reading Score'] = scores_school_type['Average Reading Score'].astype("float64")
scores_school_type['Average Math Score'] = scores_school_type['Average Math Score'].astype("float64")
scores_school_type['% Passing Math'] = scores_school_type['% Passing Math'].astype("float64")
scores_school_type['% Passing Reading'] = scores_school_type['% Passing Reading'].astype("float64")
scores_school_type['% Overall Passing'] = scores_school_type['% Overall Passing'].astype("float64")

scores_school_type_group = scores_school_type.groupby('School Type')

scores_school_type_group[['Average Math Score', 'Average Reading Score', '% Passing Math', '% Passing Reading', '% Overall Passing']].mean()
```
